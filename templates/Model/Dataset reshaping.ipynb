# Load the Dataset

```python
import pandas as pd

metadata_train = r"C:\Users\cyril\Pictures\SEM 8\CAPSTONE\Works\Deepfake_Dataset\train_sample_videos\metadata.json"
meta_train = pd.read_json(metadata_train)
```

```python
import pandas as pd

# Assuming df is an existing DataFrame. If not, initialize it:
df = pd.DataFrame()

for column in meta_train:
    print(f"Processing column: {column}")
    data = {}  # Initialize a new dictionary for each column
    data["filename"] = column
    data["label"] = meta_train[column][0]  # Assuming you meant "label" instead of "lebal"
    #print(data)
    data_df = pd.DataFrame(data, index=[0])  # Convert the dictionary to a DataFrame
    df = pd.concat([df, data_df], axis=0)  # Concatenate the DataFrames

# Reset the index of the final DataFrame
df = df.reset_index(drop=True)
```

## Moving into Real and Fake directory

```python
import pandas as pd
import os
import shutil

# Load the CSV file
df = pd.read_csv(r'C:\Users\cyril\Pictures\SEM 8\CAPSTONE\Works\Dataset\final_dataset_train_csv.csv')
path = r"C:\Users\cyril\Pictures\SEM 8\CAPSTONE\Works\Deepfake_Dataset\train_sample_videos"

# Define the paths to the 'fake' and 'real' directories
fake_dir = r"C:\Users\cyril\Pictures\SEM 8\CAPSTONE\Works\Dataset\New_Dataset\fake"
real_dir = r"C:\Users\cyril\Pictures\SEM 8\CAPSTONE\Works\Dataset\New_Dataset\real"

# Function to move files
def move_files(df):
    for index, row in df.iterrows():
        file_name = row['filename']
        label = row['label']
        # Define source path
        src_path = os.path.join(path, file_name)
        # Define destination path based on label
        if label == 'FAKE':
            dest_path = os.path.join(fake_dir, file_name)
        elif label == 'REAL':
            dest_path = os.path.join(real_dir, file_name)
        else:
            print(f"Unknown label {label} for file {file_name}")
            continue
        # Move the file
        shutil.move(src_path, dest_path)

# Move the files
move_files(df)
```
