```python
import os
import cv2
from moviepy.editor import VideoFileClip

path = r"C:\Users\cyril\Pictures\SEM 8\CAPSTONE\Works\Dataset\final\test"
newpath = r"C:\Users\cyril\Pictures\SEM 8\CAPSTONE\Works\Dataset\Audio"

for files in os.listdir(path):
        if not os.path.exists(os.path.join(newpath,files)):
            os.mkdir(os.path.join(newpath,files))
        for file in os.listdir(os.path.join(path,files)):

            video = VideoFileClip(os.path.join(path, files, file))
            audio = video.audio
            audio.write_audiofile((os.path.join(newpath, files, file)).replace(".mp4", ".wav"))

```
